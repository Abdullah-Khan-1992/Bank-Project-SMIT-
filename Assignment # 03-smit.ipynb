1. Check Discount Eligibility
Write a program to check if a customer is eligible for a discount. If the total purchase is
more than $100, apply a 10% discount and display the final price. Otherwise, display the
total price as it is.

```python
# Dictionary of items and their prices
items = {"apple": 20.5, "banana": 30.8, "bread": 35.5, "milk": 40.5, "eggs": 40.0, "butter": 60.0, "cheese": 70.5, "juice": 30.0, "rice": 40.0, "pasta": 70.0,"fish": 80.0}


cart = []
total_price = 0

print("\nAdd items to your cart (type 'done' to finish):")
while True:
    item_name = input("Enter item name: ").lower()
    if item_name == "done":
        break
    elif item_name in items:
        cart.append(item_name)
        total_price += items[item_name]
    else:
        print("Item not available. Please try again.")

# Apply discount if more than 5 items are purchased
if total_price >= 100:
    discount = total_price * 0.10
    final_price = total_price - discount
    print("\nCongratulations! You bought more than 100 rupees and earned a 10% discount!")
    print(f"Discount Amount: ${discount:.2f}")
else:
    final_price = total_price
    print("\nYou bought less than 100 rupees. No discount applied.")

# Display summary
print("\nPurchase Summary:")
print(f"Items Purchased: {', '.join([item.capitalize() for item in cart])}")
print(f"Total Items: {len(cart)}")
print(f"Total Price: ${total_price:.2f}")
print(f"Final Price After Discount (if any): ${final_price:.2f}")
```

##7. Income Tax
If a person&#39;s annual income is above $50,000, they pay 20% tax. Otherwise, they pay
10%. Calculate and display the tax amount.

```python
# Function to calculate income tax
def calculate_income_tax(annual_income):
    if annual_income > 50000:
        tax_rate = 0.20  # 20% tax for income above $50,000
        print("Annual income above $50,000: 20% tax applied.")
    else:
        tax_rate = 0.10  # 10% tax for income $50,000 or below
        print("Annual income $50,000 or below: 10% tax applied.")
    tax_amount = annual_income * tax_rate
    return tax_amount

# Main Program
try:
    # Input for annual income
    annual_income = float(input("Enter the annual income: $"))

    if annual_income < 0:
        print("Invalid input! Annual income cannot be negative.")
    else:
        # Calculate tax amount
        tax_amount = calculate_income_tax(annual_income)

        # Display result
        print(f"The calculated tax amount is: ${tax_amount:.2f}")

except ValueError:
    print("Invalid input! Please enter a valid numeric value for the income.")
```

##8. Tax Bracket
Write a program to categorize a person into tax brackets:
 Income &lt; $30,000: &quot;Low Tax&quot;
 $30,000 ≤ Income &lt; $100,000: &quot;Medium Tax&quot;
 Income ≥ $100,000: &quot;High Tax&quot;

```python
# Function to determine tax bracket
def determine_tax_bracket(annual_income):
    if annual_income < 30000:
        bracket = "Low Tax"
    elif 30000 <= annual_income < 100000:
        bracket = "Medium Tax"
    else:
        bracket = "High Tax"
    return bracket

# Main Program
try:
    # Input for annual income
    annual_income = float(input("Enter the annual income: $"))

    if annual_income < 0:
        print("Invalid input! Annual income cannot be negative.")
    else:
        # Determine tax bracket
        tax_bracket = determine_tax_bracket(annual_income)

        # Display result
        print(f"The tax bracket for an income of ${annual_income:.2f} is: {tax_bracket}")

except ValueError:
    print("Invalid input! Please enter a valid numeric value for the income.")
```

##13. Tiered Discounts
Apply discounts based on the total price:
 $0–$50: No discount.
 $50–$100: 10% discount.
 Over $100: 20% discount.

```python
# Function to calculate price after tiered discounts
def calculate_price_with_tiered_discount(total_price):
    if total_price <= 50:
        discount_rate = 0.00  # No discount
        print("Total price $0-$50: No discount applied.")
    elif 50 < total_price <= 100:
        discount_rate = 0.10  # 10% discount
        print("Total price $50-$100: 10% discount applied.")
    else:
        discount_rate = 0.20  # 20% discount
        print("Total price over $100: 20% discount applied.")

    discount_amount = total_price * discount_rate
    final_price = total_price - discount_amount
    return final_price

# Main Program
try:
    # Input for total price
    total_price = float(input("Enter the total price: $"))

    if total_price < 0:
        print("Invalid input! Total price cannot be negative.")
    else:
        # Calculate final price after discount
        final_price = calculate_price_with_tiered_discount(total_price)

        # Display result
        print(f"The final price after discount is: ${final_price:.2f}")

except ValueError:
    print("Invalid input! Please enter a valid numeric value for the total price.")
```

```python
# Function to check attendance eligibility
def check_attendance_eligibility(attendance_percentage):
    if attendance_percentage >= 75:
        print("Eligible to take the exam")
    else:
        print("Not eligible to take the exam")

# Get user input for the student's attendance percentage
attendance_percentage = float(input("Enter the student's attendance percentage: "))

# Check and display whether the student is eligible or not
check_attendance_eligibility(attendance_percentage)
```

25. Scholarship Eligibility
If a student&#39;s grade is &quot;A&quot; and their annual family income is below $30,000, they are
eligible for a scholarship. Otherwise, they are not.

```python
# Function to check scholarship eligibility
def check_scholarship_eligibility(grade, family_income):
    if grade == "A" and family_income < 30000:
        print("Eligible for the scholarship")
    else:
        print("Not eligible for the scholarship")

# Get user input for the student's grade and family income
grade = input("Enter the student's grade: ").strip().upper()
family_income = float(input("Enter the student's annual family income ($): "))

# Check and display scholarship eligibility
check_scholarship_eligibility(grade, family_income)
```
