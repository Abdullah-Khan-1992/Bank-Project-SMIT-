2. Write a program that accepts a list from user. Your program should reverse the content of list and
display it. Do not use reverse() method.

```python
# Function to reverse the list without using reverse() method
def reverse_list(user_list):
    reversed_list = []
    # Iterate over the list in reverse order and append to the new list
    for i in range(len(user_list) - 1, -1, -1):
        reversed_list.append(user_list[i])
    return reversed_list

# Get input from the user for the list
user_input = input("Enter the elements of the list separated by spaces: ")

# Convert the input string into a list
user_list = user_input.split()

# Call the function to reverse the list
reversed_list = reverse_list(user_list)

# Display the reversed list
print("Reversed list is:")
print(reversed_list)
```

```python
# Function to find the largest number in the list
def find_largest_number(user_list):
    # Initialize the largest number with the first element
    largest_number = user_list[0]

    # Iterate through the list to find the largest number
    for num in user_list:
        if num > largest_number:
            largest_number = num

    return largest_number

# Get input from the user for the list
user_input = input("Enter the elements of the list separated by spaces: ")

# Convert the input string into a list of integers
user_list = list(map(int, user_input.split()))

# Call the function to find the largest number
largest_number = find_largest_number(user_list)

# Display the largest number
print(f"The largest number in the list is: {largest_number}")
```

6. Write a program that reads a string from the user containing a date in the form mm/dd/yyyy. It
should print the date in the form March 12, 2021.

```python
# Function to convert the date from mm/dd/yyyy to Month Day, Year
def convert_date(date_str):
    # Month names
    month_names = ["January", "February", "March", "April", "May", "June",
                   "July", "August", "September", "October", "November", "December"]

    # Split the input string into month, day, and year
    month, day, year = date_str.split('/')

    # Convert the month (integer) to the corresponding month name
    month_name = month_names[int(month) - 1]

    # Print the formatted date
    return f"{month_name} {int(day)}, {year}"

# Get input from the user for the date
date_input = input("Enter a date in mm/dd/yyyy format: ")

# Call the function to convert and display the date
formatted_date = convert_date(date_input)
print(f"The formatted date is: {formatted_date}")
```

like this :

```python
# Function to find and display the sum of each row of a matrix
def sum_of_rows(matrix):
    for i, row in enumerate(matrix, start=1):
        row_sum = sum(row)  # Calculate the sum of the current row
        print(f"Sum of row {i} = {row_sum}")

# Input the matrix size
m = int(input("Enter the number of rows (m): "))
n = int(input("Enter the number of columns (n): "))

# Input the elements of the matrix
matrix = []
print("Enter the elements of the matrix row by row:")

for i in range(m):
    row = list(map(int, input(f"Enter row {i+1}: ").split()))
    matrix.append(row)

# Call the function to calculate and display the sum of each row
sum_of_rows(matrix)
```

9. Write a program to add two matrices of size n x m.

```python
# Function to add two matrices
def add_matrices(matrix1, matrix2, n, m):
    # Initialize a result matrix with zeroes
    result = [[0 for _ in range(m)] for _ in range(n)]

    # Add corresponding elements of both matrices
    for i in range(n):
        for j in range(m):
            result[i][j] = matrix1[i][j] + matrix2[i][j]

    return result

# Input the size of the matrices
n = int(input("Enter the number of rows (n): "))
m = int(input("Enter the number of columns (m): "))

# Input the first matrix
print("Enter elements of the first matrix:")
matrix1 = []
for i in range(n):
    row = list(map(int, input(f"Enter row {i+1}: ").split()))
    matrix1.append(row)

# Input the second matrix
print("Enter elements of the second matrix:")
matrix2 = []
for i in range(n):
    row = list(map(int, input(f"Enter row {i+1}: ").split()))
    matrix2.append(row)

# Add the two matrices
result_matrix = add_matrices(matrix1, matrix2, n, m)

# Display the result matrix
print("The sum of the two matrices is:")
for row in result_matrix:
    print(row)

```

10. Write a program to multiply two matrices

```python
# Function to multiply two matrices
def multiply_matrices(matrix1, matrix2, p, q, r):
    # Initialize a result matrix with zeroes of size p x r
    result = [[0 for _ in range(r)] for _ in range(p)]

    # Multiply the matrices
    for i in range(p):
        for j in range(r):
            # Compute the dot product of the i-th row of matrix1 and the j-th column of matrix2
            for k in range(q):
                result[i][j] += matrix1[i][k] * matrix2[k][j]

    return result

# Input the dimensions of the matrices
p = int(input("Enter the number of rows of matrix A (p): "))
q = int(input("Enter the number of columns of matrix A and rows of matrix B (q): "))
r = int(input("Enter the number of columns of matrix B (r): "))

# Input the first matrix (matrix A)
print("Enter elements of the first matrix (A):")
matrix1 = []
for i in range(p):
    row = list(map(int, input(f"Enter row {i+1}: ").split()))
    matrix1.append(row)

# Input the second matrix (matrix B)
print("Enter elements of the second matrix (B):")
matrix2 = []
for i in range(q):
    row = list(map(int, input(f"Enter row {i+1}: ").split()))
    matrix2.append(row)

# Multiply the matrices
result_matrix = multiply_matrices(matrix1, matrix2, p, q, r)

# Display the result matrix
print("The result of the multiplication is:")
for row in result_matrix:
    print(row)
```

2. Write a program that reads a string from keyboard and display:
* The number of uppercase letters in the string
* The number of lowercase letters in the string
* The number of digits in the string
* The number of whitespace characters in the string

```python
# Function to count uppercase, lowercase, digits, and whitespace characters
def count_characters(input_string):
    uppercase_count = 0
    lowercase_count = 0
    digit_count = 0
    whitespace_count = 0

    # Loop through each character in the string
    for char in input_string:
        if char.isupper():
            uppercase_count += 1
        elif char.islower():
            lowercase_count += 1
        elif char.isdigit():
            digit_count += 1
        elif char.isspace():
            whitespace_count += 1

    return uppercase_count, lowercase_count, digit_count, whitespace_count

# Get input from the user
user_input = input("Enter a string: ")

# Call the function to count the characters
uppercase, lowercase, digits, whitespace = count_characters(user_input)

# Display the results
print(f"Number of uppercase letters: {uppercase}")
print(f"Number of lowercase letters: {lowercase}")
print(f"Number of digits: {digits}")
print(f"Number of whitespace characters: {whitespace}")
```

For example if the user enters the string 'HELLO' then new string would be 'OELLH'

```python
# Function to swap the first and last characters of a string
def swap_first_and_last(input_string):
    if len(input_string) <= 1:
        return input_string  # If the string length is 1 or empty, return the string as is
    else:
        # Swap the first and last characters
        swapped_string = input_string[-1] + input_string[1:-1] + input_string[0]
        return swapped_string

# Get input from the user
user_input = input("Enter a string: ")

# Call the function to swap the characters
new_string = swap_first_and_last(user_input)

# Display the new string
print(f"The new string after swapping first and last characters is: {new_string}")
```

7. A palindrome is a string that reads the same backward as forward. For example, the words dad,
madam and radar are all palindromes. Write a programs that determines whether the string is a
palindrome.
Note: do not use reverse() method

```python
# Function to check if a string is a palindrome
def is_palindrome(input_string):
    # Remove spaces and convert to lowercase for a case-insensitive check
    input_string = input_string.replace(" ", "").lower()

    # Compare characters from the start and end
    for i in range(len(input_string) // 2):
        if input_string[i] != input_string[len(input_string) - 1 - i]:
            return False  # If any character doesn't match, it's not a palindrome

    return True  # If all characters match, it's a palindrome

# Get input from the user
user_input = input("Enter a string: ")

# Check if the string is a palindrome
if is_palindrome(user_input):
    print(f"'{user_input}' is a palindrome.")
else:
    print(f"'{user_input}' is not a palindrome.")
```

9. Write a program in python that accepts a string to setup a passwords. Your entered password must
meet the following requirements:

The password must be at least eight characters long.
It must contain at least one uppercase letter.
It must contain at least one lowercase letter.
It must contain at least one numeric digit.
Your program should should perform this validation.

```python
import re

# Function to validate the password
def validate_password(password):
    # Check if the password is at least 8 characters long
    if len(password) < 8:
        return "Password must be at least 8 characters long."

    # Check if the password contains at least one uppercase letter
    if not re.search(r'[A-Z]', password):
        return "Password must contain at least one uppercase letter."

    # Check if the password contains at least one lowercase letter
    if not re.search(r'[a-z]', password):
        return "Password must contain at least one lowercase letter."

    # Check if the password contains at least one digit
    if not re.search(r'[0-9]', password):
        return "Password must contain at least one digit."

    return "Password is valid!"

# Get input from the user
user_password = input("Enter your password: ")

# Validate the password
result = validate_password(user_password)

# Display the result
print(result)
```

15. Write a Python program that prompts the user to enter a base number and an
exponent, and then calculates the power of the base to the exponent. The program
should not use the exponentiation operator (**) or the math.pow() function. The
program should handle both positive and negative exponents.

```python
# Ask the user to input the base number and the exponent
base = float(input("Enter the base number: "))
exponent = int(input("Enter the exponent: "))

# Initialize the result variable
result = 1

# Check if the exponent is positive or negative
if exponent >= 0:
    # Calculate the power for positive exponent
    for i in range(exponent):
        result *= base  # Multiply the base 'exponent' times
else:
    # Calculate the power for negative exponent
    for i in range(abs(exponent)):
        result *= base  # Multiply the base 'abs(exponent)' times
    result = 1 / result  # For negative exponents, return the reciprocal of the result

# Display the result
print(f"The result of {base} raised to the power of {exponent} is: {result}")
```

```python
# Ask the user to input their weight (in kilograms) and height (in meters)
weight = float(input("Enter your weight in kilograms: "))
height = float(input("Enter your height in meters: "))

# Calculate the BMI
bmi = weight / (height * height)

# Classify the BMI into categories
if bmi < 18.5:
    print(f"Your BMI is {bmi:.2f}. You are Underweight.")
elif 18.5 <= bmi <= 24.9:
    print(f"Your BMI is {bmi:.2f}. You have Normal weight.")
elif 25 <= bmi <= 29.9:
    print(f"Your BMI is {bmi:.2f}. You are Overweight.")
else:
    print(f"Your BMI is {bmi:.2f}. You have Obesity.")
```

13. Write a program that prompts the user to input a year and determine whether the
year is a leap year or not.
Leap Years are any year that can be evenly divided by 4. A year that is evenly divisible by
100 is a leap year only if it is also evenly divisible by 400. Example:
1992 Leap Year
2000 Leap Year
1900 NOT a Leap Year
1995 NOT a Leap Year

```python
# Input year from the user
year = int(input("Enter a year: "))

# Check if the year is a leap year
if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print(f"{year} is a Leap Year.")
else:
    print(f"{year} is NOT a Leap Year.")
```

14. Write a program that prompts the user to input number of calls and calculate the
monthly telephone bills as per the following rule:
Minimum Rs. 200 for up to 100 calls.
Plus Rs. 0.60 per call for next 50 calls.
Plus Rs. 0.50 per call for next 50 calls.
Plus Rs. 0.40 per call for any call beyond 200 calls.

```python
# Input the number of calls from the user
calls = int(input("Enter the number of calls made: "))

# Base bill for up to 100 calls
bill = 200

# Calculate additional charges based on the number of calls
if calls > 100 and calls <= 150:
    bill += (calls - 100) * 0.60  # Rs. 0.60 per call for next 50 calls
elif calls > 150 and calls <= 200:
    bill += (50 * 0.60) + (calls - 150) * 0.50  # Rs. 0.50 per call for next 50 calls
elif calls > 200:
    bill += (50 * 0.60) + (50 * 0.50) + (calls - 200) * 0.40  # Rs. 0.40 per call beyond 200 calls

# Output the calculated bill
print(f"Your monthly telephone bill is Rs. {bill:.2f}")
```
